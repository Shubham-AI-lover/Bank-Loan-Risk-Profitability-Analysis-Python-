```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('loan_cleaned.csv')
```

```python
df.head()
```

### Loans Issued per Year

```python
loans_per_year = df.groupby('issue_year').size()
```

```python
loans_per_year
```

```python
loans_per_year.plot(figsize=(8,5))
plt.title("Loans Issued per Year")
plt.xlabel("Year")
plt.ylabel("Number of Loans")
plt.show()
```

**Upward trend  
Accelerating growth**  

Loan issuance shows strong year-over-year growth, indicating rapid expansion of lending activity.

### Default Rate Trend

```python
default_trend = df.groupby('issue_year')['default_flag'].mean()
```

```python
default_trend
```

```python
default_trend.plot(figsize=(8,5))
plt.title("Default Rate Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Default Rate")
plt.show()
```

 **2007  2009**  
Default rate decreases significantly  
Risk is improving  

**2009  2010**  
Roughly stable  

**2010  2011**  
Default rate increases  
Risk starts worsening  

Risk **improved initially**, reached a low around 2009, and then **worsened slightly** afterward.

Default rates declined sharply from 2007 to 2009, indicating improving portfolio quality, followed by a mild increase after 2009, suggesting slight risk deterioration.

### Profit Trend

```python
profit_trend = df.groupby('issue_year')['Profit'].mean()
```

```python
profit_trend
```

```python
profit_trend.plot(figsize=(8,5))
plt.title("Average Profit per Loan Over Time")
plt.xlabel("Year")
plt.ylabel("Average Profit")
plt.show()
```

2007  slightly negative  
2008  more negative  
2009  strong positive  
2010  slightly lower than 2009 but still positive  
2011  highest positive value  

**Profitability is improving over time (after an early dip).**  

Average profit per loan declined between 2007 and 2008, but improved significantly from 2009 onward, indicating strengthening portfolio profitability over time.  

## Final Business Conclusions

- Lending volume shows strong long-term growth, indicating platform expansion  
- Default rates declined initially and increased slightly in later years, suggesting emerging risk pressure  
- Average profit per loan improved significantly after 2009 and remains positive  
- Higher-risk loan grades exhibit higher default rates but also higher average profits, reflecting a riskreturn trade-off  
- Mid-grade borrowers offer a balanced combination of risk and profitability
